```python
#Importando as bibliotecas
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
#Leitura do arquivo
videogames = pd.read_csv('vgsales.csv')
```

```python
#Renomeando colunas
videogames.columns=['Rankings', 'Nome', 'Plataforma', 'Ano', 'Genero', 'Editora', 'Vendas America do Norte', 'Vendas EUA', 'Vendas Japão', 'Outras vendas', 'Vendas Global']
```

```python
#Exibição dos dataset
videogames.head(10)
```

 <h3 align="center">  Quantidade de jogos lançados apartir do ano 2000 </h3>



```python
top = videogames[['Genero', 'Vendas America do Norte']].head(10).set_index('Genero').sort_values('Vendas America do Norte', ascending= True)
top.plot(kind='barh', figsize =(11,7), grid = False, color = 'darkred', legend = False )
plt.title('As plataformas mais vendidas na America do Norte')
plt.show()
```

```python
topeua = videogames[['Genero', 'Vendas EUA']].head(10).set_index('Genero').sort_values('Vendas EUA', ascending= True)
top.plot(kind='barh', figsize =(11,7), grid = False, color = 'green', legend = False )
plt.title('As plataformas mais vendidas na Europa')
plt.xlabel('Total de vendas')
plt.show()
```

```python
#Criando grafico 
videogames['Plataforma'].value_counts().head(10).plot(kind='bar', figsize=(11,5), grid = False, rot = 0, color = 'green')
#Colocando alguns nomes no grafico
plt.title('Os 10 videogames com mais titulos lançados') #Definindo o titulo
plt.xlabel('Plataformas') #Nomeando o eixo X
plt.ylabel('Quantidade de jogos lançados') #Nomeando o eixo Y
plt.show() #Exibindo o gráfico
```

```python
#Criando grafico 
videogames['Editora'].value_counts().head(5).plot(kind='bar', figsize=(11,5), grid = False, rot = 0, color = 'Blue')
#Colocando alguns nomes no grafico
plt.title('As 5 publisher com mais titulos') #Definindo o titulo
plt.xlabel('Publisher') #Nomeando o eixo X
plt.ylabel('Quantidade de jogos lançados') #Nomeando o eixo Y
plt.show() #Exibindo o gráfico
```

```python
#Os mais vendidos no mundo
top_10_vendidos = videogames[['Nome', 'Vendas Global']].head(10).set_index('Nome').sort_values('Vendas Global', ascending= True)
top_10_vendidos.plot(kind='barh', figsize =(11,7), grid = False, color = 'purple', legend = False )
#Colocando o titulos e outros nomes
plt.title('Os 10 jogos mais vendidos no mundo')
plt.xlabel('Total de vendas(milhões de doláres)')
plt.show()
```
